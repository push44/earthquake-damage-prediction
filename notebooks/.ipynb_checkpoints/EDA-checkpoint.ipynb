```python
def describe_columns(df, labels):
    
    import matplotlib.pyplot as plt
    import numpy as np
    import seaborn as sns
    
    ### Overview ###
    
    print("Shape:", df.shape, "\n")
    print("Unique dtypes:", df.dtypes.unique(), "\n")
    
    for dtype in df.dtypes.unique():
        print(f"{dtype}: # of columns", df.select_dtypes(include=dtype).shape[1])
    print("\n")
    
    for col in df.columns:
        null_sum = df[col].isnull().sum()
        if null_sum>0:
            print(col, "% missing values:", df[col].isnull().sum()/df.shape[1])
    
    dummy_df = df.copy(deep=True)
    dummy_df["damage"] = labels.values
    
    ### Categorical Columns ###
    
    categorical_levels = []
    label_variance = []
    
    for col in df.select_dtypes(include="object").columns:
        categorical_levels.append(len(df[col].unique()))
        
        label_variance.append(dummy_df.groupby(by=col)["damage"].var().values)
        
    fig, [ax1, ax2, ax3] = plt.subplots(nrows=3, ncols=1, figsize=(18,13), sharex=True)
    
    ax1.bar(x=range(df.select_dtypes(include="object").shape[1]), height=categorical_levels)
    ax1.set_ylabel("Number of categorical\nlevels", fontsize=15)
    
    ax2.plot(list(map(lambda col_vals: np.mean(col_vals), label_variance)), "o--")
    ax2.set_ylabel("Label variance\ngroup by column", fontsize=15)
    
    ax3.boxplot(label_variance, positions=range(df.select_dtypes(include="object").shape[1]))
    ax3.set_ylabel("Label variance\ngrouped by column", fontsize=15)
    ax3.set_xticks(range(df.select_dtypes(include="object").shape[1]))
    ax3.set_xticklabels(df.select_dtypes(include="object").columns, rotation=90, fontsize=13)
    
    fig.suptitle("Categorical Features", fontsize=16)
    
    ### Numerical Columns ###
    
    fig, ax4 = plt.subplots(figsize=(13,13))
    sns.heatmap(df.select_dtypes(exclude="object").drop(["building_id"], axis=1).corr(), cbar=False, annot=True, fmt=".1f", ax=ax4)
    fig.suptitle("Numerical Features", fontsize=16)
    
    cnt=0
    label_variance = []
    for col in df.select_dtypes(exclude="object").drop(["building_id"], axis=1).columns:
        if len(df[col].unique())==2:
            label_variance.append([col]+list(dummy_df.groupby(by=col)["damage"].var().values))
            cnt+=1
    
    print("# Binary columns:", cnt)
    
    fig, ax5 = plt.subplots(figsize=(13,5))
    ax5.plot(list(map(lambda col_vals: col_vals[1], label_variance)), "o--", alpha=0.3)
    ax5.plot(list(map(lambda col_vals: col_vals[2], label_variance)), "o--", alpha=0.3)
    ax5.plot(list(map(lambda col_vals: np.mean(col_vals[1:]), label_variance)), "o--")
    ax5.set_xticks(range(len(label_variance)))
    ax5.set_xticklabels(list(map(lambda col_vals: col_vals[0], label_variance)), rotation=90)
    ax5.set_ylabel("Label variance\ngrouped by binary", fontsize=15)
```

```python
describe_columns(train_values, train_labels["damage_grade"])
```
